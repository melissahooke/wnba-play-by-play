```python
import requests
import json
import pandas as pd
import re
from datetime import datetime
```

```python
year = '2020'
game_type = '2' #game_type = 1:preseason; 2:regular season; 3:all-star; 4:playoff
game_num = '001' #game_num = 1-n for preseason/regular season; series+0+number(1-5)
quarter_num = '1'
gid = '10'+game_type+year[2:4]+'00'+game_num
gid = '1022001001'
quarter_num = '1'
game_url = 'https://data.wnba.com/data/5s/v2015/json/mobile_teams/wnba/'+year+'/scores/pbp/'+gid+'_'+quarter_num+'_pbp.json'
r = requests.get(game_url)
game_data = json.loads(r.text)
```

```python
def save_new_rows(df,game_data,game_season):
    game_info = game_data['gcode']
    newRows = pd.DataFrame.from_dict(game_data['pla'])
    game_info = re.split('/',game_info)
    newRows['gid'] = gid
    newRows['year'] = year
    newRows['gameType'] = game_season
    newRows['quarter'] = int(quarter_num)
    newRows['date'] = datetime.strptime(game_info[0],'%Y%m%d')
    newRows['homeTeam'] = game_info[1][3:6]
    newRows['awayTeam'] = game_info[1][0:3]
    df = df.append(newRows)
    return df
```

```python
game_type = '1' #game_type = 1:preseason; 2:regular season; 3:all-star; 4:playoff
df = pd.DataFrame() 

#loop through game numbers
for game_num in [str(x) for x in list(range(1,300))]:
    game_num = (4-len(game_num))*'0'+game_num #game_num = 1-n for preseason/regular season; series+0+number(1-5) for playoffs
    print(game_num)
    gid = '10'+game_type+year[2:4]+'0'+game_num
    quarter_num = '1'
    game_url = 'https://data.wnba.com/data/5s/v2015/json/mobile_teams/wnba/'+year+'/scores/pbp/'+gid+'_'+quarter_num+'_pbp.json'
    r = requests.get(game_url)
    game_data = json.loads(r.text)
    try:
        game_data = game_data['g']
        df = save_new_rows(df,game_data,'preseason')
    except:
        break
        
    #loop through quarters
    for quarter_num in [str(x) for x in list(range(2,10))]:
        game_url = 'https://data.wnba.com/data/5s/v2015/json/mobile_teams/wnba/'+year+'/scores/pbp/'+gid+'_'+quarter_num+'_pbp.json'
        r = requests.get(game_url)
        game_data = json.loads(r.text)
        try:
            game_data = game_data['g']
            df = save_new_rows(df,game_data,'preseason')
        except:
            break
preseason = df
```

```python
game_type = '2' #game_type = 1:preseason; 2:regular season; 3:all-star; 4:playoff
df = pd.DataFrame() 

#loop through game numbers
for game_num in [str(x) for x in list(range(1,300))]:
    if year == '2020':
        game_num = '1'+(3-len(game_num))*'0'+game_num #game_num = 1-n for preseason/regular season; series+0+number(1-5) for playoffs
    else:
        game_num = (4-len(game_num))*'0'+game_num #game_num = 1-n for preseason/regular season; series+0+number(1-5) for playoffs
    
    gid = '10'+game_type+year[2:4]+'0'+game_num
    print(gid)
    quarter_num = '1'
    game_url = 'https://data.wnba.com/data/5s/v2015/json/mobile_teams/wnba/'+year+'/scores/pbp/'+gid+'_'+quarter_num+'_pbp.json'
    r = requests.get(game_url)
    game_data = json.loads(r.text)
    try:
        game_data = game_data['g']
        df = save_new_rows(df,game_data,'regular season')
    except:
        break
    
    #loop through quarters
    for quarter_num in [str(x) for x in list(range(2,10))]:
        game_url = 'https://data.wnba.com/data/5s/v2015/json/mobile_teams/wnba/'+year+'/scores/pbp/'+gid+'_'+quarter_num+'_pbp.json'
        r = requests.get(game_url)
        game_data = json.loads(r.text)
        try:
            game_data = game_data['g']
            df = save_new_rows(df,game_data,'regular season')
        except:
            break
regular = df
```

```python
game_type = '4' #game_type = 1:preseason; 2:regular season; 3:all-star; 4:playoff
df = pd.DataFrame() 

#loop through game numbers
for round_num in [str(x) for x in list(range(1,5))]:
    for series_num in [str(x) for x in list(range(0,5))]:
        for game_num in [str(x) for x in list(range(1,6))]:
            game_num = '0'+round_num+series_num+game_num #game_num = 1-n for preseason/regular season; series+0+number(1-5) for playoffs
            gid = '10'+game_type+year[2:4]+'0'+game_num
            print(gid)
            quarter_num = '1'
            game_url = 'https://data.wnba.com/data/5s/v2015/json/mobile_teams/wnba/'+year+'/scores/pbp/'+gid+'_'+quarter_num+'_pbp.json'
            r = requests.get(game_url)
            game_data = json.loads(r.text)
            try:
                game_data = game_data['g']
                df = save_new_rows(df,game_data,'playoffs')
            except:
                continue
                
            #loop through quarters
            for quarter_num in [str(x) for x in list(range(2,10))]:
                game_url = 'https://data.wnba.com/data/5s/v2015/json/mobile_teams/wnba/'+year+'/scores/pbp/'+gid+'_'+quarter_num+'_pbp.json'
                r = requests.get(game_url)
                game_data = json.loads(r.text)
                try:
                    game_data = game_data['g']
                    df = save_new_rows(df,game_data,'playoffs')
                except:
                    break
playoffs = df
```
